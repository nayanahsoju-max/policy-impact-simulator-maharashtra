```python

import pandas as pd
import re

file_path = r"C:\Users\HP\Downloads\DDW-2700C-08.xlsx"

AREA_COL     = 3
TRU_COL      = 4
AGE_COL      = 5
CATEGORY_COL = 6
VALUE_COL    = 7

df = pd.read_excel(file_path, skiprows=5, header=None)
df.columns = range(len(df.columns))

def extract_name(area_str, keyword):
    after = area_str.split(keyword, 1)[1].strip()
    return re.sub(r"\s+\d+$", "", after).strip().lower()

current_state    = None
current_district = None

rows = []
workers_dict = {}

for _, row in df.iterrows():
    area = str(row[AREA_COL]).strip() if pd.notna(row[AREA_COL]) else ""
    tru  = str(row[TRU_COL]).strip()  if pd.notna(row[TRU_COL])  else ""
    age  = str(row[AGE_COL]).strip()  if pd.notna(row[AGE_COL])  else ""
    category = str(row[CATEGORY_COL]).strip() if pd.notna(row[CATEGORY_COL]) else ""

    if "State -" in area:
        current_state = extract_name(area, "State -")
        current_district = None

    if "District -" in area:
        current_district = extract_name(area, "District -")

    # --- Literacy ---
    if tru == "Total" and age == "All ages":
        total    = row[6]
        literate = row[12]

        if pd.notna(total) and pd.notna(literate):
            try:
                literacy_rate = float(literate) / float(total)
            except:
                continue

            if current_state and current_district:
                rows.append([current_state, current_district, literacy_rate, total])

    # --- Workers ---
    if "Total workers" in category and tru == "Total":
        workers = row[VALUE_COL]

        if pd.notna(workers) and current_state and current_district:
            key = (current_state, current_district)
            workers_dict[key] = float(workers)

# Create dataframe
df_final = pd.DataFrame(rows, columns=["state", "district", "literacy_rate", "total_population"])

# Map workers
df_final["total_workers"] = df_final.apply(
    lambda x: workers_dict.get((x["state"], x["district"])), axis=1
)

# Compute worker ratio
df_final["worker_ratio"] = df_final["total_workers"] / df_final["total_population"]

print(df_final[["state", "district", "literacy_rate", "worker_ratio"]].to_string(index=False))
```

```python
import numpy as np

final["vulnerability_score"] = (
    0.3 * final["st_pct"] +
    0.2 * final["sc_pct"] +
    0.2 * (1 - final["literacy_rate"]) +
    0.3 * final["deprivation_score"]
)

final["vulnerability_score"] = (
    (final["vulnerability_score"] - final["vulnerability_score"].min()) /
    (final["vulnerability_score"].max() - final["vulnerability_score"].min())
)

print(final.head())
```

```python
threshold = final["vulnerability_score"].quantile(0.75)
final["high_risk"] = (final["vulnerability_score"] >= threshold).astype(int)
```
